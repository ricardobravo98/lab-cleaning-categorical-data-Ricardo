```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
customer_df = pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
customer_df
```

```python
customer_df.columns 
```

```python
numerical_df = customer_df.select_dtypes("number")
categorical_df = customer_df.select_dtypes(exclude = "number")
```

```python
numerical_df
```

```python
for column in categorical_df.columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=column, y='Total Claim Amount', data=customer_df)
    plt.show()
```

```python
for col in categorical_df:
    print(col)
    x = categorical_df[col].value_counts()
    print(x)
    print(" ")
```
